## How to measure monetization impact on engagement?

We are introducing a new metric here mDAU / DAU. It's percentage of people in DAU who have been converted into paying customers. 

Our simple model predicts that this ratio eventually becomes stable. If paying users don't increase their engagement, although they inherently have better-than-average engagement, over a long enough period of time less engaged non-payers will churn, leaving both groups with roughly the same distribution. This can be seen bellow and it happens significant time after the conversion curve starts flattening.

As it is stated in the `metrics-model`, there is some change of payer engagement over long period of time that we haven't counted in the model (yet). This can be observed as divergence of real mDAU / DAU from teretically estimated fDAU / DAU which basically flattens at one point.

Consequently, if mDAU / DAU starts going down after it has flattened, this is a clear sign that paying users are getting disappointed after a while and are negativelly affected.
